```python
import pandas as pd
import numpy as np
import random
```

```python
# Read csv
weekly_raffle_df = pd.read_csv("11.28.2020.csv")
```

```python
# Drop all rows with missing values
weekly_raffle_df = weekly_raffle_df.dropna()
```

```python
# Drop all rows with headers repeating
weekly_raffle_df = weekly_raffle_df[weekly_raffle_df.Username != "Username"]
weekly_raffle_df
```

```python
# Select only Username and Total Deposited columns in new df
clean_wr_df = weekly_raffle_df[['Username','Total Deposited']]
```

```python
# Convert Total Deposited to float number and Username to string
clean_wr_df['Total Deposited'] = clean_wr_df['Total Deposited'].str.replace('$', '').astype(float)
clean_wr_df['Username'] = clean_wr_df['Username'].astype(str)
```

```python
# Groupby Username and sum total deposited. Create Entries column for every $10 deposited and round down value
grouped_wf = clean_wr_df.groupby('Username').sum().reset_index()
grouped_wf['Entries'] = grouped_wf['Total Deposited']/10
grouped_wf['Entries'] = grouped_wf['Entries'].apply(np.floor).astype(int)
```

```python
grouped_wf.head()
```

```python
# Loop through Usernames and multiply by number of entries and store in list
names_list = []
for ind in grouped_wf.index: 
    names_list += grouped_wf['Entries'][ind] * [grouped_wf['Username'][ind]]
    
names_list
```

```python
print("The winner for this past raffle week is", random.choice(names_list))
```
