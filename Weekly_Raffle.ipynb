```python
import pandas as pd
```

```python
# Read csv
weekly_raffle_df = pd.read_csv("11.9.2020.csv")
```

```python
# Drop all rows with missing values
weekly_raffle_df = weekly_raffle_df.dropna()
```

```python
# Drop all rows with headers repeating
weekly_raffle_df = weekly_raffle_df[weekly_raffle_df.Username != "Username"]
weekly_raffle_df
```

```python
# Convert Total Deposited values to float
# weekly_raffle_df.astype(str)
weekly_raffle_df['Total Deposited'].str.replace('$', '').astype(float)
clean_wr_df = weekly_raffle_df[['Username','Total Deposited']]
```

```python
weekly_raffle_df = weekly_raffle_df.astype(str)
weekly_raffle_df.dtypes
```
